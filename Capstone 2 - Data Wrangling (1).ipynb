```python
import pandas as pd 
import numpy as np
```

```python
deaths_data = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv')
deaths_df=pd.DataFrame(deaths_data)
```

```python
print(deaths_df.head())
print(pd.DataFrame.info(deaths_df))
```

```python
deaths_df['date'] = pd.to_datetime(deaths_df['date'])
most_recent_date = deaths_df['date'].max()
deaths_summed = deaths_df[deaths_df['date']==most_recent_date]
print(deaths_summed.head(10))
```

```python
data = pd.read_csv(r'state_demographics.csv')
data_df = pd.DataFrame(data)
#print(data_df.head(2))
```

```python
ethnicitiesdata_df = data_df[['State','Ethnicities.Black Alone','Ethnicities.American Indian and Alaska Native Alone',
                  'Ethnicities.Asian Alone', 'Ethnicities.Native Hawaiian and Other Pacific Islander Alone',
                 'Ethnicities.Two or More Races','Ethnicities.Hispanic or Latino','Ethnicities.White Alone, not Hispanic or Latino']]
```

```python
#print(ethnicitiesdata_df.head(5))
```

```python
ED = ethnicitiesdata_df.sort_values('State')
ED.head(10)
```

```python
ED.columns = map(str.lower, ED.columns)
ED.head(5)
```

```python
macrodata_df = data_df[['State','Income.Median Houseold Income','Income.Persons Below Poverty Level']]
MD = macrodata_df.sort_values('State')
MD.tail(10)
```

```python
MD.columns = map(str.lower, MD.columns)
MD.head(5)
```

```python
merged_data1 = pd.merge(deaths_summed, ED, on='state', how='outer')
merged_data1.head(5)
```

```python
df = pd.merge(merged_data1, MD, on='state', how='outer')
df.tail(5)
```

```python
#clean up column names and dropping unnecessary columns
df = df.drop(df.columns[[0,2]], axis=1)
df = df.dropna()
#df.columns = df.columns.str.strip('ethnicites')
df.columns = df.columns.str.capitalize()
df.head()
```

```python
df.to_csv('capstone2_df.csv')
```
